```python
import optuna
from optuna import Trial, visualization
from optuna.samplers import TPESampler
```

```python

def objectiveXGB(trial: Trial,X,y,X_test,y_test):
    param = {
        "n_estimators" : trial.suggest_int('n_estimators', 100,1000),
        'max_depth':trial.suggest_int('max_depth', 8, 30),
        'min_child_weight':trial.suggest_int('min_child_weight', 1, 30),
        'learning_rate': 0.01,
        'tree_method': 'gpu_hist',
        'random_state': 1
    }

    model_xgb = XGBClassifier(**param)
    model_xgb.fit(X,y, verbose=False)


    score = accuracy_score(model_xgb.predict(X_test), y_test)

    return score
```

```python
def objectiveLGBM(trial: Trial,X,y,X_test,y_test):
    param = {
        "n_estimators" : trial.suggest_int('n_estimators', 300, 1500),
        # 'max_depth':trial.suggest_int('max_depth', 5, 40),
        'learning_rate': trial.suggest_categorical('learning_rate',[0.01,0.05,0.1,0.2,0.3,0.5]),

        'num_leaves':trial.suggest_int('num_leaves',20,100)

    }

    model_lgb = LGBMClassifier(**param)
    model_lgb.fit(X,y)


    score = accuracy_score(model_xgb.predict(X_test), y_test)

    return score
```

```python
# direction : score 값을 최대 또는 최소로 하는 방향으로 지정
study = optuna.create_study(direction='maximize',sampler=TPESampler())

# n_trials : 시도 횟수 (미 입력시 Key interrupt가 있을 때까지 무한 반복)
study.optimize(lambda trial : objectiveXGB(trial,log_X_train_final,y_train,log_X_valid_final,y_valid), n_trials=50)
print('Best trial: score {},\nparams {}'.format(study.best_trial.value,study.best_trial.params))
```

```python
# direction : score 값을 최대 또는 최소로 하는 방향으로 지정
study_lgbm = optuna.create_study(direction='maximize',sampler=TPESampler())

# n_trials : 시도 횟수 (미 입력시 Key interrupt가 있을 때까지 무한 반복)
study_lgbm.optimize(lambda trial : objectiveLGBM(trial,log_X_train_final,y_train,log_X_valid_final,y_valid), n_trials=50)
print('Best trial: score {},\nparams {}'.format(study_lgbm.best_trial.value,study_lgbm.best_trial.params))
```

##MLP model

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset,DataLoader,TensorDataset
from torch.autograd import Variable
import torch.optim as optim
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

torch.manual_seed(777)
if device == 'cuda':
    torch.cuda.manual_seed_all(777)
#gpu 사용 위해
```

```python
log_X_train_final_DL=torch.tensor(log_X_train_final.values,dtype=torch.float32).to(device)
y_train_DL=torch.tensor(y_train.values,dtype=torch.float32).to(device)

log_X_valid_final_DL=torch.tensor(log_X_valid_final.values,dtype=torch.float32).to(device)
y_valid_DL=torch.tensor(y_valid.values,dtype=torch.float32).to(device)
```

```python
train_dataset = TensorDataset(log_X_train_final_DL, y_train_DL)
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True, drop_last=True)
```

```python
class Model_loan_status(nn.Module):
    def __init__(self, input_size):
        super(Model_loan_status, self).__init__()

        self.fc1 = nn.Linear(input_size, 256)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(256, 64)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(64, 32)
        self.relu3 = nn.ReLU()
        # self.fc4 = nn.Linear(64,16)
        # self.relu4 = nn.ReLU()
        self.output_layer = nn.Linear(32, 1)

    def forward(self, x):
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        x = self.relu3(self.fc3(x))
        # x = self.relu4(self.fc4(x))
        x = self.output_layer(x)
        x = torch.sigmoid(x)
        return x
```

```python
input_size = len(log_X_train.columns)
model = Model_loan_status(input_size).to(device)

criterion = nn.BCELoss().to(device)

#최적화 담당
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
num_epoch=200

for epoch in range(num_epoch):
    cost = 0.0

    for x, y in train_dataloader:
        x = x.to(device)
        y = y.to(device)

        output = model(x)
        loss = criterion(output, y)

        optimizer.zero_grad()
        loss.backward()#역전파
        optimizer.step()#가중치 업데이트

        cost += loss

    cost = cost / len(train_dataloader)

    print(f'Epoch {epoch+1:4d}/{num_epoch}, Train Loss: {cost}')
```

```python
from sklearn.metrics import accuracy_score
```

```python
#예측
with torch.no_grad():
    model.eval()
    predictions = model(log_X_valid_final_DL)
    predicted_labels = (predictions > 0.5).int()

# 정확도 계산
accuracy = accuracy_score(y_valid_DL.cpu().numpy(), predicted_labels.cpu().numpy())#gpu에서 numpy는 사용 불가능이므로
print(f'Accuracy: {accuracy:}')
```
